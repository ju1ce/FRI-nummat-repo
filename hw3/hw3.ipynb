```julia
function sekantna(f, x1, x2, maxit=100, tol = 1e-5)
      
    x = 0
    
    for i=1:maxit
        fx1 = f(x1)
        fx2 = f(x2)
        
        x = x2 - (fx2 * (x2-x1))/(fx2-fx1)
        
        #display(fx1)
        #display(fx2)
        #display(x)
        #display(abs(f(x)))
        #display("-------")
        
        x1 = x2
        x2 = x #0.9*x2+0.1*x
        
        if abs(f(x)) < tol
            return x,i
        end
        
    end
    
    return x,maxit
        
end
```

```julia
func(x) = cos(x)-0.5

sekantna(func,0.2,0.3)
```

```julia
xs = []

for i=0:0.1:4pi
    x,it = sekantna(func,i,i+0.1)
    if x > 0 && x < 4pi
        xs = vcat(xs,x)
    end
end

```

```julia
plot(func,0:0.1:4pi)
scatter!(xs,zeros(size(xs)))
```
